Seaborn:
---------------
- Seaborn is a visualization library in python for plotting statistical graphs
- Designed to work with dataframe objects in pandas
- It contains default attractive style
- It provides high-level interface for drawing attractive and informative statistical graphs

```python
import seaborn as sns
```

#### Types of plots
- color palette
    - It is an interface to generate few colors in seaborn
- plotting with categorical data
- Joint plot
- Pairplot
- Heatmap

```python
sns.color_palette()
```

```python
sns.palplot(sns.color_palette())
```

```python
sns.palplot(sns.color_palette("deep"))
```

```python
sns.palplot(sns.color_palette("deep",n_colors=7))
```

```python
sns.palplot(sns.dark_palette("purple"))
```

```python
sns.palplot(sns.light_palette("purple"))
```

```python
sns.palplot(sns.dark_palette("purple",reverse = True))
```

```python
# categorical plotting
```

```python
sns.get_dataset_names()
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris.shape
```

```python
iris.isnull().sum()
```

```python
iris["species"].value_counts()
```

```python
sns.catplot(x = "species",y="sepal_length",data = iris)
```

```python
sns.catplot(x = "species", y = "sepal_length",data = iris, kind = "box")
```

```python
sns.catplot(data = iris,kind = "box",orient = "h")
```

```python
sns.catplot(x = "species",y = "petal_width",data= iris, kind ="violin")
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
titanic.shape
```

```python
titanic.isnull().sum()
```

```python
sns.catplot(x = "sex",y = "survived",data = titanic, kind = "bar")
```

```python
sns.catplot(x = "sex",y = "survived",data = titanic, hue = "class",kind = "bar")
```

```python
# joint plot - combination of two plots and by default it have scatter and histogram
sns.jointplot(x = "petal_length",y = "sepal_width",data = iris)
```

```python
sns.jointplot(x = "petal_length",y = "sepal_width",data = iris,kind = "kde",color = "g")
```

#### Pair plot
- it displays multiple plots at a time in a singe graph
- pairwise relationship
- by default it returns the scatter plot

```python
sns.pairplot(iris)
# 1st row --> (sl,sl)(sw,sl)(pl,sl)(pw,sl)
```

```python
sns.pairplot(iris, vars = ["sepal_length","sepal_width"])
```

```python
# heat Map - it is used to find the correlation two columns in data
```

```python
t = sns.load_dataset("tips")
t.head()
```

```python
t.corr()
```

```python
sns.heatmap(t.corr(),annot = True)
```

```python
sns.heatmap(t.corr(),annot = True,center=0,linewidth = 5,linecolor = "white")
```

```python
sns.heatmap(t.corr(),annot = True, cmap = "YlGnBu")
```
