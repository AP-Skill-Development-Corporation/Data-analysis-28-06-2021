### Lineplot
* plt.plot()

```python
x_axis=[10,12,14,16]
y_axis=[5,6,7,8]
plt.plot(x_axis,y_axis)
```

```python
x_axis=[10,12,14,16]
y_axis=[5,6,7,8]
plt.plot(x_axis,y_axis,label="LINE")
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.title('Line Graph')
plt.legend(loc='lower right')
```
