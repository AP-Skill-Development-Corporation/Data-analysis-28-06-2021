```python
# Creating DataFrame by using pandas series

import pandas as pd
s=pd.Series([100,200,300,400],index=[1,2,3,4])
df=pd.DataFrame(s)
df
```

```python
# Creating Data Frame from dictionary

d={"a":pd.Series([10,20,30,40],index=[1,2,3,4]),
  "b":pd.Series([10,20,30],index=[1,2,3]),
   "c":pd.Series([20,30,45],index=[2,3,5])  
  }
df2=pd.DataFrame(d)
df2
```

```python
df2.columns=list("xyz")
df2
```

### Rows having the index values and columns having meaningful name/info about data

```python
#Create one Dataframe having squares and cubes

import numpy as np

f=[[i**2 for i in range(10,21)],[j**3 for j in range(10,21)]]
pd.DataFrame(f,index=('squares','cubes'),columns=np.arange(10,21))
```

```python
data=[{"squares":num**2,"cubes":num**3}  for num in range(10,21)]
df1=pd.DataFrame(data,index=np.arange(10,21))
df1
```

```python
s=[1,2,3,4]

d={'a':pd.Series([i**2 for i in s],index=s),
  'b':pd.Series([i**3 for i in s],index=s)
  }

df=pd.DataFrame(d)
df
```

### Combining data

** merging/concat/append different dataframes into a single dataframe`

```python
df1
```

```python
pd.concat([df1,df1],axis=1)   #rows-> axis=0 , columns= axis=1
```

```python
df1.append(df1)  #row appending
```

```python
d1=pd.DataFrame({"emp":pd.Series(["a","b","c","d"],index=[1,2,3,4]),
                 "yr":pd.Series([2000,2001,2002,2003],index=[1,2,3,4])
                })
print(d1)
```

```python
d2= pd.DataFrame({"emp":pd.Series(["a","b","c","e"],index=[1,2,3,4]),
                 "dept":pd.Series(["hr","op","mg","s"],index=[1,2,3,4])
                 })
d2
```

```python
pd.merge(d1,d2)   #based on emp names
```

```python
pd.merge(d1,d2,how="left")   #left join
```

```python
pd.merge(d1,d2,how="right")  #right join
```
