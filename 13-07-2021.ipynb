```python
plt.plot(x1,x1+10,color='r',linestyle='--')   #red color dashed line
```

```python
x2=np.linspace(10,30,50)
x2
```

```python
#marker - circle, size 20, markerfacecolor=cyan

plt.plot(x2,np.sin(x2),color='r',marker='o',markersize=20,markerfacecolor="c")
```

```python
# marker -diamond,  size=20, markerfacecolor=gray

plt.plot(x2,np.cos(x2),color='b',marker='D',markersize=20,markerfacecolor='g')
```

### Bar Chart

* plt.bar()

```python
year=['2019 Dec','2020 Jan','2020 March','2020 July']
GoldPrice=[1000,3000,5000,5300]
plt.bar(year,GoldPrice,color='r')
```

```python
year=['2019 Dec','2020 Jan','2020 March','2020 July']
Goldrates=[1000,3000,5000,5300]
Silverrates=[500,300,400,1000]
plt.bar(year,Goldrates,width=0.33)
plt.bar(year,Silverrates,width=0.33)

```

```python
plt.bar(year,Goldrates,width=0.33)
plt.bar(year,Silverrates,width=0.55)
```

```python
x=np.arange(5)
x
```

```python
plt.bar(x,[90,89,76,78,88],width=0.33)
plt.bar(x+0.33,[23,90,92,44,30],width=0.33)
```

```python
x=[1,2,3,4,5]
y=[10,55,23,44,67]
plt.bar(x,y,color=['r','c','g','b','m'],edgecolor='blue')
```

```python
x=np.random.randint(100,1000,100)
x
```

```python
y=np.random.randint(500,1500,100)
y
```

```python
plt.scatter(x,y)
```

```python
# marker as *, color as magenta

plt.scatter(x,y,marker='*',color='m')
```

```python
# marker as Diamond and green color

plt.scatter(x,y,marker="D",color='g')
```

```python
# marker as circle and yellow color

x=[12,23,45,66,78]
y=[10,20,30,40,59]
plt.scatter(x,y,marker='o',color='y')
```

```python
x=np.arange(500)
x
```

```python
plt.scatter(x,np.sin(x),marker='+',color='b')
plt.scatter(x,np.cos(x),marker='*',color='m')
```
